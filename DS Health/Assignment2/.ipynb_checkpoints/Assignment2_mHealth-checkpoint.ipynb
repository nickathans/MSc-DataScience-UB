In terms of the sensor data, we observe similar patterns of movement in the arm sensors as in the leg and torso sensors. In general, the arm sensors are seem to have different patterns of movement compared to the leg and torso sensors, as the arms may move more independently from the rest of the body. $$\textbf{Precision} = \frac{TP}{TP + FP}, $$

Finally, the $f_{1} score$ is a weighted average of precision and recall. The  $f_{1} score$  is used to get a measure of both types of failures.

# First Approach

#### Train and test sets that contain shuffled samples from each user. So the model will train on data from every user and predict the activities from every user in the test set.

The shape of X_train would be (number of samples in the training set, number of features), and the shape of y_train would be (number of samples in the training set,).

In the first approach, where shuffled samples from each user are used in the train and test sets, the data size would be the total number of segments across all activities and all subjects. From the dataset description, we know that there are 19 activities, 8 subjects, and 60 segments for each activity and subject. Therefore, the total number of segments would be:

19 activities x 8 subjects x 60 segments = 9120 segments

The features are extracted from each segment of sensor data, and 235 features are extracted per segment.

These features are a combination of statistical moments (mean, standard deviation, skewness, and kurtosis), autocorrelation, and peaks in the discrete Fourier transform (DFT). Since each 5-second segment contains 45 sensors (5 units x 9 sensors), and we are computing the first four statistical moments for each sensor, we have a total of 45 x 4 = 180 features. Additionally, we are also considering the first n peaks of the discrete Fourier transform of each segment. So if we choose n to be 55, then we have an additional 55 features, making a total of 180 + 55 = 235 features for each segment. Since we have 480 segments per activity per subject, the shape of the data is (9120, 235).

So if we follow the first approach (shuffled samples from each user), the shape of X_train would be (9120, 235) and the shape of y_train would be (9120,).

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier


# Load data
activities = ['a01', 'a02', 'a03', 'a04', 'a05', 'a06', 'a07', 'a08', 'a09','a10', 'a11', 'a12', 'a13', 'a14', 'a15', 'a16', 'a17', 'a18', 'a19']
subjects = ['p1', 'p2', 'p3', 'p4', 'p5', 'p6', 'p7', 'p8']
units = ['T', 'RA', 'LA', 'RL', 'LL']
sensors = ['acc', 'gyro', 'mag']
n_segments = 60
segment_length = 125
n_features = 4 + 9 + 10  # Moments, autocorrelation and DFT peaks

data = []
labels = []
for activity in activities:
    for subject in subjects:
        for segment in range(1, n_segments+1):
            filename = path + f'{activity}/{subject}/s{segment:02d}.txt'
            df = pd.read_csv(filename, header=None)
            
            ####    Extract features    ####
            
            # mean, variance,skewness, and the kurtosis
            moments = df.apply(lambda x: pd.Series(x).describe()[1:5].values).values.reshape(-1)
            
            # autocorrelation
            autocorr = df.apply(lambda x: pd.Series(x).autocorr(lag=5)).values
            
            # discrete Fourier transform 
            dft_peaks = np.abs(np.fft.rfft(df)).max(axis=0)[1:11]
            
            # Concatenate features and append to data
            features = np.concatenate((moments, autocorr, dft_peaks))
            data.append(features)
            
            # Assign label
            label = activities.index(activity)
            labels.append(label)

data = np.array(data)
labels = np.array(labels)
```

### Optimize model Hyperparameters

After defining the pipeline, a grid of hyperparameters to search over is specified in param_grid. The hyperparameters to optimize include the number of principal components to retain (pca__n_components), the number of decision trees in the random forest classifier (clf__n_estimators), and the maximum depth of the decision trees (clf__max_depth).

The GridSearchCV function from scikit-learn is used to perform a grid search over the hyperparameters specified in param_grid. The cv=5 argument specifies a 5-fold cross-validation, where the data is split into 5 parts and the model is trained and evaluated 5 times, each time using a different part of the data as the validation set. The n_jobs=-1 argument specifies that the grid search should use all available CPUs to speed up computation. Finally, the cv.fit(X_train, y_train) call fits the model to the training data and searches over the hyperparameters to find the best combination of hyperparameters that yields the highest cross-validation accuracy.

The best hyperparameters are printed out using cv.best_params_.

```python
# Optimize model hyperparameters
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA()),
    ('clf', RandomForestClassifier())
])
param_grid = {
    'pca__n_components': [10, 20, 30],
    'clf__n_estimators': [50, 100, 200],
    'clf__max_depth': [5, 10, None]
}
cv = GridSearchCV(pipe, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)
cv.fit(X_train, y_train)
print('Best parameters:', cv.best_params_)
```

### Cross Validation

Cross-validation score is a measure of a machine learning model's performance on data that it has not been trained on. Cross-validation is a technique used to evaluate machine learning models by dividing the dataset into multiple subsets, called folds, and training the model on a combination of these folds while testing on the remaining fold. This process is repeated multiple times with different folds used for training and testing to obtain an estimate of the model's performance on unseen data.

The cross-validation score is the average score across all folds and is used to evaluate the model's generalization performance. It is a more reliable estimate of the model's performance than the training score, which measures how well the model fits the training data. The cross-validation score provides an estimate of how well the model is likely to perform on new, unseen data.

The learning_curve function uses cross-validation to generate training and cross-validation scores for the model at different training set sizes. The cross-validation score is used to plot the learning curve and evaluate the model's performance.

```python
import matplotlib.pyplot as plt

# Cross Validate model's performance by analyze learning curves
train_sizes = np.linspace(0.1, 1.0, 10)
train_sizes, train_scores, test_scores = learning_curve(
    cv.best_estimator_, X_train, y_train, cv=5, n_jobs=-1, train_sizes=train_sizes)
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plot the learning curves
plt.plot(train_sizes, train_mean, label='Training score')
plt.plot(train_sizes, test_mean, label='Cross-validation score')

# Fill the area between the curves
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1)
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1)

# Add labels and legend
plt.xlabel('Training set size')
plt.ylabel('Score')
plt.title('Learning curves')
plt.legend()

# Show the plot
plt.show()
```

This code generates a plot with the training score and cross-validation score on the y-axis and the training set size on the x-axis. The shaded areas represent the standard deviation of the scores for each training set size. The legend shows the labels for the training score and cross-validation score.

In this case, the learning curve for the first approach shows that the training score is steady around 1.0, which means that the model is able to perfectly fit the training data. This is expected because the model has access to all the data from all users in the training set. However, the cross-validation score starts from 0.96 and gradually increases as the training set size increases. This suggests that the model is able to generalize well to new data.

As the training set size increases, the model has more examples to learn from, which leads to an improvement in the model's ability to generalize. However, after a certain point, the performance of the model starts to plateau, which means that adding more data to the training set may not lead to a significant improvement in performance.

Overall, the learning curve for the first approach indicates that the model is not overfitting the data and is able to generalize well to new data, which is supported by the high testing scores.
